```python
import pandas as pd
import numpy as np
```

```python
DATASET_PATH = 'adult.data'

columns = ['age',
           'workclass',
           'fnlwgt',
           'education',
           'education_num',
           'marital_status',
           'occupation', 
           'relationship', 
           'race',
           'sex',
           'capital_gain',
           'capital_loss', 
           'hours_per_week',
           'native_country',
           'class']
```

```python
df = pd.read_csv(DATASET_PATH, header=None)
df.columns = columns
```

```python
category_columns = ['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country', 'class']
for column in category_columns:
    df[column] = df[column].astype('category')
    df[column] = df[column].cat.codes
```

```python
continues_columns = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']
categorical_columns = ['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country']
```

```python
for column in categorical_columns:
    tempdf = pd.get_dummies(df[column], prefix=column)
    
    df = pd.merge(
        left=df,
        right=tempdf,
        left_index=True,
        right_index=True,
    )
    
    df = df.drop(columns=column)
```

```python
continues_columns = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']

for column in continues_columns:
    df[column] = (df[column] - df[column].mean()) / df[column].std()
```

```python
df.to_csv('training_data.csv', index=False)
```
